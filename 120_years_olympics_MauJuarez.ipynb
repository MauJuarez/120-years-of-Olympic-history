## 10.- Find the Ratio of male and female athletes participated in all olympic games.

Creamos un dataframe auxiliar donde solo tendremos el dato de sexo, nombre y el juego realizado

```python
datos_genero = dataset[['Name','Sex','Games']]
datos_genero
```

Ahora agruparemos por sexo, en diferentes columna, luego calcularemos el total de participantes, porcentaje de participacion femenina y masculina, y el ratio de participacion femenina respecto a la masculina, por cada juego 

Recordemos que para hacer la agrupación por varios campos, es poniendo esos campos en una lista groupby([campo1, campo2 ,,,]).

size: para contar la cantidad de filas en cada grupo resultante


unstack: para transformar los resultados en un DataFrame con las columnas "games", "M" y "F"

Además fillna(0) le dara el valor de cero, cuando en alguno no halla datos disponibles


```python
grupo_genero = datos_genero.groupby(['Games','Sex'])['Name'].size().unstack().reset_index()
grupo_genero['Total de participantes'] =grupo_genero['M']+grupo_genero['F'].fillna(0)
grupo_genero['Porcentaje Femenino'] =grupo_genero['F'].fillna(0)/grupo_genero['Total de participantes']*100
grupo_genero['Porcentaje Masculino'] =grupo_genero['M']/grupo_genero['Total de participantes']*100
grupo_genero['Porcentaje F/M'] =grupo_genero['F'].fillna(0)/grupo_genero['M']*100
grupo_genero
```

## 16.-Identify which country won the most gold, most silver and most bronze medals in each olympic games.

```python
medals_by_country = grupo_medallas_2
medals_by_country
```

Las siguientes lineas nos permiten obtener un dataframe por tipo de medalla, donde obtendremos el juego y el pais que las gano.

```python
max_gold = medals_by_country.groupby('Games').apply(lambda x: x[x['Gold'] == x['Gold'].max()])[['Games', 'NOC', 'Gold']]
max_silver = medals_by_country.groupby('Games').apply(lambda x: x[x['Silver'] == x['Silver'].max()])[['Games', 'NOC', 'Silver']]
max_bronze = medals_by_country.groupby('Games').apply(lambda x: x[x['Bronze'] == x['Bronze'].max()])[['Games', 'NOC', 'Bronze']]
```

```python
max_gold.reset_index(drop=True, inplace=True)
max_silver.reset_index(drop=True, inplace=True)
max_bronze.reset_index(drop=True, inplace=True)
```

Mostramos los dataframe de cada una de las categorias, por ejemplo, para las de plata.

```python
max_silver
```

Unimos los dataframes generados para cada categoria con pd.merge

```python
max_gs =pd.merge(max_gold,max_silver, on = 'Games')
max_gsb = pd.merge(max_gs,max_bronze, on = 'Games')
max_gsb
```

## 17.-Identify which country won the most gold, most silver, most bronze medals and the most medals in each olympic games.

Haremos algo muy parecido a lo anterior, pero ahora incluyendo totales.

```python
medals_by_country_tot =medals_by_country
medals_by_country_tot['Total']=medals_by_country_tot['Gold'].fillna(0)+medals_by_country_tot['Silver'].fillna(0)+medals_by_country_tot['Bronze'].fillna(0)
medals_by_country_tot
```

Ahora procedemos de manera exactamente igual que antes pero incluyendo a la columna total.

```python
max_gold = medals_by_country.groupby('Games').apply(lambda x: x[x['Gold'] == x['Gold'].max()])[['Games', 'NOC', 'Gold']]
max_silver = medals_by_country.groupby('Games').apply(lambda x: x[x['Silver'] == x['Silver'].max()])[['Games', 'NOC', 'Silver']]
max_bronze = medals_by_country.groupby('Games').apply(lambda x: x[x['Bronze'] == x['Bronze'].max()])[['Games', 'NOC', 'Bronze']]
max_medals = medals_by_country.groupby('Games').apply(lambda x: x[x['Total'] == x['Total'].max()])[['Games', 'NOC', 'Total']]
max_gold.reset_index(drop=True, inplace=True)
max_silver.reset_index(drop=True, inplace=True)
max_bronze.reset_index(drop=True, inplace=True)
max_medals.reset_index(drop=True, inplace=True)
max_gs =pd.merge(max_gold,max_silver, on = 'Games')
max_gsb = pd.merge(max_gs,max_bronze, on = 'Games')
max_gsbt= pd.merge(max_gsb,max_medals, on = 'Games')
max_gsbt
```
